```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
import statsmodels.api as sm
from statsmodels.formula.api import ols
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_excel('childmortality.xlsx')
m = df.dropna()
```

```python
mortalidade = m.iloc[0:,[0,208]].dropna()
mortalidade
```

```python
fertility = pd.read_excel('totalfertility.xlsx')
f=fertility.dropna()
```

```python
fertility_ano = f.iloc[0:,[0,208]]
fertility_ano
```

```python
life = pd.read_excel('lifeexpectancy.xlsx')
l = life.dropna()
```

```python
life_ano = l.iloc[0:,[0,208]].dropna()
life_ano
```

```python
life_ano['fertility'] = fertility_ano[2007]
life_ano['expectativa']=life_ano[2007]
```

```python
life_ano['mortality']=mortalidade[2007]
life_ano
```

```python
x=life_ano[['expectativa','fertility']].copy()
y=life_ano['mortality'].copy()
x=sm.add_constant(x)
```

```python
est1=sm.OLS(y,x).fit()
est1.summary()

#life_ano.plot(kind='scatter', x='mortality', y='expectativa')
```

```python
xx1,xx2 = np.meshgrid(np.linspace(x.expectativa.min(),x.expectativa.max(),100),(np.linspace(x.fertility.min(),x.fertility.max(),100)))
z1=est1.params[0]+est1.params[1]*xx1+est1.params[2]*xx2
fig=plt.figure(figsize=(12,8))
ax=Axes3D(fig,azim=-115,elev=15)
surf=ax.plot_surface(xx1,xx2,z1,cmap=plt.cm.RdBu_r,alpha=0.6,linewidth=0)
resid=y-est1.predict(x)
ax.scatter(x[resid>=0].expectativa,x[resid>=0].fertility,y[resid>=0],color='black',alpha=1.0,facecolor='none')
ax.scatter(x[resid<0].expectativa,x[resid<0].fertility,y[resid<0],color='black',alpha=1.0)
ax.set_xlabel('Expectativa de Vida')
ax.set_ylabel('Fertilidade')
ax.set_zlabel('Mortalidade')
```
