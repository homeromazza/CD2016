Outro ponto importante sobre a divisão da educação, é que a porcentagem de engajamento tecnológico em aposentados que só fizeram Ensino Fundamental e aposentados que só fizeram Ensino Médio é relativamente parecida, o que expressa que o grau de escolaridade dentro do grupo dos aposentados com afinidade tecnlógica não é algo relevante.
Não obstante, visto que a vida acadêmica dos aposentados não é um fator decisivo sobre o engajamento tecnológico, analisando a renda podemos obter mais algumas informações sobre o perfil do aposentado que ainda usa a tecnologia. A partir de uma análise se torna expressivo a relação entre o engajamento tecnológico e a renda, sendo que, no geral, a população com engajamento tecnológico possui uma renda melhor que a população sem engajamento tecnológico, com menores diferenças de rendas entre os indivíduos de cada população e um valor salarial relativamente maior como um todo. Não obstante disso, vale ressaltar que existe uma diferença bem significante entre os engajados tecnologicamente da população que fez só até o Ensino Fundamental e os não engajados dessa mesma população.
Portanto, após todos esses dados e análises, se torna um pouco mais claro o comportamento/perfil dos aposentados que não usam tecnologia e dos que de fato utilizam dela. É visto que, no geral,  o aposentado com engajamento tecnológico tem maior probabilidade de possuir um salário melhor e inclusive, com uma certeza bem maior, viver em populaçao urbana. Porém, visto de uma esfera menos detalhista, se pode observar que a quantidade de aposentados com engajamento tecnológico é excessivamente baixa e isso  não possui uma relação muito consistente com o grau de escolaridade,e assim se torna viável formular a hipótese de que o baixo engajamento tecnológico é algo característico desta geração estudada, e não de fatores mais especifícos sobre essa cada indivíduo dessa população.
